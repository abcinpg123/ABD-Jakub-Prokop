Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

1. $x>-1$ oraz $x<1$

2. $x>-6$ oraz $x<6$

3. $x>0$ oraz $x<5$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def f(x):
  return x**2+5
```

```python
a1 = np.linspace(-1, 1, 100)
a2 = np.linspace(-6, 6, 100)
a3= np.linspace(0, 5, 100)

fig, ax = plt.subplots()
ax.plot(a1, f(a1), label="f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(-1, 1)")
ax.legend()
plt.show()

fig2, ax2 = plt.subplots()
ax2.plot(a2, f(a2), label="f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(-6, 6)")
ax2.legend()
plt.show()

fig3, ax3 = plt.subplots()
ax3.plot(a3, f(a3), label="f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(0, 5)")
ax3.legend()
plt.show()
```
