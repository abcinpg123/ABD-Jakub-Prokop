```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return x**2+5
```

```python
a1 = np.linspace(-1, 1, 100)
a2 = np.linspace(-6, 6, 100)
a3= np.linspace(0, 5, 100)

plt.plot(a1, f(a1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(-1, 1)")
plt.show()
plt.legend(a1, "a1")

plt.plot(a2, f(a1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(-6, 6)")
plt.show()

plt.plot(a3, f(a1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("(0, 5)")
plt.show()
```
