```python
# 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

import pandas as pd

tb = pd.read_csv("../Original data/tb.csv")

cols = list(tb.columns)
# tb = tb.dropna(axis=1, how='all', subset=cols[2:])
tb.head(10)
```

```python
new_cols = []
for col in tb.columns[3:]:
    string = col.split('_')[2]
    sex = ""
    age = ""

    if string[0]=='m':
        sex += "Male"
    elif string[0]=='f':
        sex += "Female"

    age = string[1:]
    if age == "u":
        age = "age_unknown"
    elif age == "65":
        age = "65+"
    elif len(age) == 2:
        age = age[0] + '-' + age[1]
    elif len(age) == 3:
        age = age[0:1] + '-' + age[1:3]
    elif len(age) == 4:
        age = age[0:2] + '-' + age[2:4] 

    new_cols.append(sex + ' ' + age)
# print(new_cols)
# print(cols)

transform_cols = {}
for i in range(len(new_cols)):
    transform_cols[cols[i+3]] = new_cols[i]

print(transform_cols)
tb = tb.rename(columns=new_cols)

tb.head(10)
```
