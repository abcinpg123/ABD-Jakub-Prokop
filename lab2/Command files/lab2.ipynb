```python
# 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

import pandas as pd
import os

tb = pd.read_csv("../Analysis data/tb.csv")

tb.head(10)
```

```python
tb = tb.drop(['new_sp'], axis=1) # usuniÄ™cie kolumny 'new_sp' - nie do koÅ„ca wiadomo co ona zawiera, byÄ‡ moÅ¼e wszystkie przypadki?
cols = list(tb.columns)

tb = tb.dropna(axis=0, how='all', subset=cols[2:]) # usuniÄ™cie wierszy zawiarajÄ…cych jedynie wartoÅ›ci NaN w kolumnach innych niÅ¼ "iso2" i "year"
tb.head(10000)
```

```python
# zmiana nazw kolumn na bardziej czytelne

new_cols = ["Country", "Year"]
for col in tb.columns[2:]:
    string = col.split('_')[2]
    sex = ""
    age = ""

    if string[0]=='m':
        sex += "Male"
    elif string[0]=='f':
        sex += "Female"

    age = string[1:]
    if age == "u":
        age = "age_unknown"
    elif age == "65":
        age = "65+"
    elif len(age) == 2:
        age = age[0] + '-' + age[1]
    elif len(age) == 3:
        age = age[0:1] + '-' + age[1:3]
    elif len(age) == 4:
        age = age[0:2] + '-' + age[2:4] 

    new_cols.append(sex + ' ' + age)
# print(new_cols)
# print(cols)

tb.columns = new_cols

tb.head(10)
```

```python
# przeniesienie danych z kaÅ¼dego kraju do osobnej tabeli; stworzenie sÅ‚ownika w formacie { Kraj(iso) : tabela }

data_dict = {}

for iso in pd.unique(tb.Country):
    data_dict[iso] = tb.set_index('Country').filter(like=str(iso), axis=0).set_index('Year')

data_dict['PL'].head()
```

```python
# usuniÄ™cie z tabel kolumn zawierajÄ…cych wyÅ‚Ä…cznie brakujÄ…ce dane (wartoÅ›ci NaN)

for iso, table in data_dict.items():
    table.dropna(axis=1, how='all', inplace=True)
    data_dict[iso] = table

data_dict['PL'].head(100)
```

```python
# zapis do pliku

new_file_path = '../Analysis data/processed_tb.csv'

if os.path.isfile(new_file_path):            # usuÅ„ plik jeÅ›li istnieje
    os.remove(new_file_path)

with open(new_file_path, 'a', newline="") as f:          # otwÃ³rz plik i zapisz dane
    for iso, table in data_dict.items():
        f.write("Country ISO: " + str(iso))
        f.write("\n")
        table.to_csv(f, encoding = "utf-8-sig")
        f.write("\n")
        f.write("\n")
```
