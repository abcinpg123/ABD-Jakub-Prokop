```python
import pandas as pd

tb = pd.read_csv("../Original data/tb.csv")

cols = tb.columns
# tb = tb.dropna(axis=1, how='all', subset=cols[2:])
tb.head(10)
```

```python
new_cols = []
for col in tb.columns[3:]:
    string = col.split('_')[2]
    sex = ""
    age = ""

    if string[0]=='m':
        sex += "Male"
    elif string[0]=='f':
        sex += "Female"

    age = string[1:]
    print(age)
    if len(age) == 2:
        age = age[0] + '-' + age[1]
    elif len(age) == 3:
        age = age[0] + '-' + age[1:2]
    elif len(age) == 4:
        age = age[0:2] + '-' + age[2:3] 
    elif age == "u":
        age = "age_unknown"

    new_cols.append(sex + ' ' + age)
print(new_cols)

tb.head(10)
```
